```python
import pandas as pd

# Leggi il file CSV
df = pd.read_csv('merged_output.csv')
```

```python
df
print(df)
```

```python

partiti = ["FratelliDItalia", "PD", "PCD", "MV5S", "Lega", "ForzaItalia", "StatiUnitiDeuropa", "avs"]
party_votes = df[partiti].sum()
print(party_votes)
total_votes = party_votes.sum()
party_percentages = (party_votes / total_votes) * 100
print(party_percentages)
```

```python
import matplotlib.pyplot as plt

# Save the pie chart for party percentages
plt.figure(figsize=(8, 8))
plt.pie(party_percentages, labels=party_votes.index, autopct='%1.1f%%')
plt.title('Party Percentages')
plt.savefig('party_percentages.png')

# Save the pie chart for party votes
plt.figure(figsize=(8, 8))
plt.pie(party_votes, labels=[f"{party} ({votes})" for party, votes in party_votes.items()], autopct='%1.1f%%')
plt.title('Party Votes')
plt.savefig('party_votes.png')

```

```python
import pandas as pd

df = pd.read_csv('merged_output_quartieri.csv')


grouped_df = df.groupby('Quartiere').agg({'Sezione': 'first', 'FratelliDItalia': 'sum', 'PD': 'sum', 'PCD': 'sum', 'Lega': 'sum', 'MV5S': 'sum', 'StatiUnitiDeuropa': 'sum', 'avs': 'sum', 'ForzaItalia': 'sum'})


```

```python
grouped_df
```

```python
partiti =[ "FratelliDItalia", "PD", "PCD", "MV5S", "Lega", "ForzaItalia", "StatiUnitiDeuropa", "avs"]
```

```python
# Calculate the sum of total votes for each row
grouped_df['Total Votes'] = grouped_df[partiti].sum(axis=1)

grouped_df

```

```python
import os
import matplotlib.pyplot as plt


# Create the "grafici" folder if it doesn't exist
if not os.path.exists("grafici"):
    os.makedirs("grafici")

# Iterate over each quartiere
for quartiere in grouped_df.index:
    # Get the party votes for the quartiere
    quartiere_votes = grouped_df.loc[quartiere, partiti]
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(quartiere_votes, labels=[f"{party} ({votes})" for party, votes in quartiere_votes.items()], autopct='%1.1f%%')
    plt.title(f"Party Votes - {quartiere}")
    
    # Save the chart in the "grafici" folder with the name "nomequartiere.png"
    plt.savefig(f"grafici/{quartiere}.png")
    plt.close()
```
